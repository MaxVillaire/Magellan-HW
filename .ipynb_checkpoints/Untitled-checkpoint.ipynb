```python
'''
In this file I will fix the formatting of the previously generated numbers file
'''

import re

def fix_file(in_file, out_file):
    r = re.compile('^\(?(\d{3})\)?[\s.-]?(\d{3})[\s.-]?(\d{4})$')
    with open(in_file,'r') as reader:
        with open(out_file,'w') as writer:
            for line in reader:
                tmp = r.match(line)
                writer.write(tmp.group(1)+tmp.group(2)+tmp.group(3)+'\n')
```

```python
def find_matching(in_file):
    fix_file(in_file,''in_file)
    nums = []
    with open(in_file, 'r') as reader:
        for line in reader:
            nums.append(int(line))
    prev = ''
    dups = []
    for n in sorted(nums):
        if n == prev:
            dups.append(n)
        prev = n
    print('Duplicates removed:')
    print(dups)
    if dups: nums.remove(dups)
    with open('sorted_'+in_file, 'w') as writer:
        for n in nums:
            writer.write('({0}){1}-{2}'.format(n[:3],n[3:6],n[6:10]))
    return dups
```

```python
for n in range(1,101):
    fix_file('phones{0}.txt'.format(n),'phones{0}.txt'.format(n-1))
```

```python
s = 0
for n in range(0,100):
    s = bool(find_matching('phones{0}.txt'.format(n)))
print(s/100.0)
```
