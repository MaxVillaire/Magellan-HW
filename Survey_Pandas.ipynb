```python
import csv
from collections import defaultdict
import pandas as pd
import numpy as np
```

```python
def read_results():
    table = defaultdict(dict)
    with open('survey_results_public.csv','r') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            if row['ProgramHobby'] != 'NA' and row['BoringDetails'] != 'NA':
                row_data = row['ProgramHobby']
                col_data = row['BoringDetails']
                try:
                    table[row_data][col_data] += 1
                except:
                    table[row_data][col_data] = 1
        
        return table
```

```python
print(read_results())
```
